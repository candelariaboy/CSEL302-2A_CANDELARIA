```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('BankCustomerData.csv')
```

```python
df_dummies = pd.get_dummies(df,drop_first=True)
```

```python
df_dummies["subscribed"] = (df["age"]>30).astype(int)
```

```python
x = df_dummies.drop(["age", "subscribed"],axis=1)
y = df_dummies["subscribed"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=30)
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
```

```python
model = LogisticRegression()
model.fit(x_train_scaled,y_train)
```

```python
y_pred = model.predict(x_test_scaled)
```

```python
accuracy = accuracy_score(y_test,y_pred)
bank_matrix = confusion_matrix(y_test,y_pred)
bank_report = classification_report(y_test,y_pred)

print(f"Accuracy: {accuracy}")
print("Confusion Matrix")
print(bank_matrix)
print("Classification Reports: ")
print(bank_report)
```
